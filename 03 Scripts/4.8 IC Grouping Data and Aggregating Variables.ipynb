# 03. Calculate Mean of Order Number

```python
# Use groupby and agg functions to calculate mean of user_order_number
df_ords_prods_derived.groupby('department_id').agg({'order_number' : ['mean']})
```

Notes: applying the groupby and agg. functions to the entire dataframe showed more department_ids compared to the subset of the dataframe.
Plus, the mean values of each department getting higher when looking at the entire dataframe.

# 04. Create Loyalty Flag for Exisiting Customers

```python
# Use groupby and transform() functions with np.max argument to determine maximum order number for each user
df_ords_prods_derived['max_order'] = df_ords_prods_derived.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
##  Check updated dataframe
df_ords_prods_derived.head(15)
```

```python
## Use loc function to make loyalty flag column with given criteria
df_ords_prods_derived.loc[df_ords_prods_derived['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_derived.loc[(df_ords_prods_derived['max_order'] <= 40) & (df_ords_prods_derived['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_derived.loc[df_ords_prods_derived['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
## Check frequency of values in new loyalty flag column
df_ords_prods_derived['loyalty_flag'].value_counts(dropna = False)
```

```python
df_ords_prods_derived.head()
```

# 05. Look at Spending Habits of Loyalty Groups

```python
# Use groupby and agg() to determine basic statisitcs of spending habits by loyalty group
df_ords_prods_derived.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

```python
### Note: The average price (mean) of a product purchased by a new customer is about $3 more than that of a loyal customer.
```

## Note: we can see that the maximum value is very far from the mean and medianm so we can just treat them as missing values using this function

```python
df_ords_prods_derived.loc[df_ords_prods_derived['prices'] >100, 'prices'] = np.nan
```

```python
df_ords_prods_derived['prices'].max()
```

# 06 Create a spending flag for each user based on the average price across all their orders

```python
## Use transform() to determine the mean price of products purchased by each user
df_ords_prods_derived['mean_product_price'] = df_ords_prods_derived.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# View relevant columns 
df_ords_prods_derived[['user_id', 'mean_product_price']].head(15)
```

```python
## Create spending_flag column with given criteria using loc() function
df_ords_prods_derived.loc[df_ords_prods_derived['mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_ords_prods_derived.loc[df_ords_prods_derived['mean_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
## Check frequency of new spending_flag column
df_ords_prods_derived['spending_flag'].value_counts(dropna = False)
```

```python
## View relevant columns of updated dataframe
df_ords_prods_derived[['user_id', 'mean_product_price', 'spending_flag']].head(20)
```

# 07  Create an order frequency flag that marks the regularity of a userâ€™s ordering behavior according to the median in the â€œdays_since_prior_orderâ€ column.

```python
# Use groupby and transform to determine median number of days between orders for each user
df_ords_prods_derived['median_days_between_orders'] = df_ords_prods_derived.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# View relevant columns of updated dataframe
df_ords_prods_derived[['user_id', 'days_since_prior_order', 'median_days_between_orders']].head(20)
```

```python
# Create order frequency flag based on given criteria using loc function
df_ords_prods_derived.loc[df_ords_prods_derived['median_days_between_orders'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
df_ords_prods_derived.loc[(df_ords_prods_derived['median_days_between_orders'] <= 20) & (df_ords_prods_derived['median_days_between_orders'] > 10), 'order_frequency'] = 'Regular customer'
```

```python
df_ords_prods_derived.loc[df_ords_prods_derived['median_days_between_orders'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
# Check value counts of new order_frequency column
df_ords_prods_derived['order_frequency'].value_counts(dropna = False)
```

There seems to be rows that were not labeled. One explanation for this could be that there are some users who have only ever made one order.



```python
# View relevant columns of updated data frame
df_ords_prods_derived[['user_id', 'days_since_prior_order', 'median_days_between_orders', 'order_frequency']].head(30)
```
