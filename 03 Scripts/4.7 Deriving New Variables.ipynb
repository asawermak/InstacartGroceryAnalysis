# 1a.Create "price_label" column

```python
## Create a subset of first million rows
df = df_ords_prods_merged
```

```python
df['prices'].max()
```

```python
## Define a function for price labels

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
## Apply a function for price range. 
df['price_range'] = df.apply(price_label, axis=1)
```

# 1b. Create "busiest_day" column.

```python
# Rename "order_dow" column to "order_day_of_week."
df.rename(columns = {'order_dow' : 'orders_day_of_week'}, inplace = True)
```

```python
# Find frequency of orders_day_of_week.
df['orders_day_of_week'].value_counts(dropna = False)
```

```python
result = []

for value in df["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Create new column in df_ords_prods_merged to view results in context.
df['busiest_day'] = result
```

```python
df.head()
```

```python
df['busiest_day'].value_counts(dropna = False)
```

# 2. Update "busiest day" column to "busiest days" to identify order day of the week as "Busiest days," "Slowest days," or "Regularly busy."

```python
# Create column “busiest_days" from "orders_day_of_week" to identify two busiest days (0, 1), slowest days (4, 3), or regularly busy (all other days).

result_2 = []

for value in df["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Regularly busy")
```

```python
result_2
```

# 3. Check values of new "busiest days" column for accuracy with observations in markdown format.

```python
# Create new 'busiest_days' column in ords_prods_merge to view results in context.

df['busiest_days'] = result_2
```

```python
df['busiest_days'].value_counts(dropna = False)
```

# 4. Create new column "busiest_period_of_day" to identify time periods “Most orders,” “Average orders,” and “Fewest orders.”

```python
# Check value counts in "order_hour_of_day" column.
df['order_hour_of_day'].value_counts()
```

### Note: 
The value counts listed above will be split into equal thirds for the following labels: "Most orders" = 10, 11, 14, 13, 15, 12, 16, and 9;
"Fewest orders" = 23, 6, 0, 1, 5, 2, 3, and 4; "Average orders" = all remaining values in order_hour_of_day.


```python
df.head()
```

#  5. Print frequency of "busiest_period_of_day" column.

```python
def find_busiest_period(row):
    if row["order_hour_of_day"] in [10, 11, 14, 15, 13, 12, 16, 9]:
        result = ("Most orders")
    elif row["order_hour_of_day"] in [23, 6, 0, 1, 5, 2, 4, 3]:
        result = ("Fewest orders")
    else:
        result = ("Average orders")
    return result

df['busiest_period_of_day'] = df.apply(find_busiest_period, axis=1)
```

```python
## Print value counts in "busiest_period_of_day" column.
df['busiest_period_of_day'].value_counts(dropna = False)
```
